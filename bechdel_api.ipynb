```python
count = 0

# create an empty dataframe
df_new = pd.DataFrame(columns=['title', 'year', 'rating', 'dubious', 'imdbid', 'id', 'submitterid', 'date', 'visible'])

# call from the latest to the earliest movie, in case the website crashes
for imdbid in df['imdbid'][-1::-1]:
    # Fetch the data from the URL
    response = requests.get('http://bechdeltest.com/api/v1/getMovieByImdbId?imdbid=' + str(imdbid))
    
    if response.status_code == 200:
        # Convert the JSON response to a pandas Series
        series = pd.Series(response.json())
        # Convert the Series to DataFrame and append it to df_new
        df_new = pd.concat([df_new, series.to_frame().T], ignore_index=True)

        if count == 4300:
            break
        count += 1
         
        
    else:
        print(f"Failed to fetch data for IMDb ID {imdbid}: {response.status_code}")

# Save the DataFrame to a CSV file
df_new.to_csv('C:\Projects Shata\Bechdel-Test\Bechdel_detailed_reversed.csv', index=False)

# Print the number of movie data entries saved to the CSV file
print(f"Number of movie data entries saved to the CSV file: {len(df_new)}")

```

```python
# reverse 
df_new = df_new.iloc[::-1].reset_index(drop=True)
df_new.to_csv('C:\Projects Shata\Bechdel-Test\Bechdel_detailed.csv') #, index=False)
```

```python
# CHECK DATA INTEGRITY

df = pd.read_csv('C:\Projects Shata\Bechdel-Test\Bechdel.csv', index_col=0)
df_to_check = pd.read_csv('C:\Projects Shata\Bechdel-Test\Bechdel_detailed_limited.csv', index_col=0)
df_to_check
```

```python

def check_dataset(df_to_check):

    movieFlag = False
    lengthFlag = len(df_to_check) == len(df)
    print(f'Same number of rows? {lengthFlag}, should be {len(df)} movies, got {len(df_to_check)} movies')
    if lengthFlag:
        movieFlag = True
        for i in range(len(df)):
            if df['imdbid'].iloc[i] != df_to_check['imdbid'].iloc[i]:
                print(i, df['imdbid'].iloc[i], df_to_check['imdbid'].iloc[i])
                movieFlag = False
                break
    print(f'Same movies? {movieFlag}')
```

```python

# check integrity and fail
check_dataset(df_to_check)
```
